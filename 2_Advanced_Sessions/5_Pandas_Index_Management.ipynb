```python
import pandas as pd

from datasets import load_dataset
import matplotlib.pyplot as plt

import numpy as np

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.set_index('job_posted_date', inplace = True)
df.head()
```

```python
df.reset_index('job_posted_date', inplace = True)
df.head()
```

```python
median_salary_pivot = df.pivot_table(values = 'salary_year_avg', index = 'company_name', aggfunc = 'median')
median_salary_pivot.sort_index(inplace = True, ascending = False)
median_salary_pivot.head(5)
```

```python
median_title_pivot = df.pivot_table(values = 'salary_year_avg', index = 'job_title_short', aggfunc = 'median')
median_title_pivot.reset_index(inplace = True)
median_title_pivot.set_index('salary_year_avg', inplace = True)
median_title_pivot.head()
```
