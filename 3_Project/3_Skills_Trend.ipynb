# How are the top skills trending for Data Analyst?

```python
import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda date: ast.literal_eval(date) if pd.notna(date) else date)
```

```python
df_us_da = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')].copy()
df_us_da['job_posted_month_num'] = df_us_da['job_posted_date'].dt.month
```

```python
df_us_da_explode = df_us_da.explode('job_skills')
```

```python
df_us_da_pivot = df_us_da_explode.pivot_table(index='job_posted_month_num', columns='job_skills', aggfunc='size', fill_value=0)
df_us_da_pivot.loc['Total'] = df_us_da_pivot.sum()
df_us_da_pivot = df_us_da_pivot[df_us_da_pivot.loc['Total'].sort_values(ascending=False).index]
df_us_da_pivot = df_us_da_pivot.drop('Total')

df_us_da_pivot
```

```python
da_total = df_us_da.groupby('job_posted_month_num').size()
```

```python
df_us_da_percent = df_us_da_pivot.div(da_total/100, axis=0)
```

```python
df_us_da_percent = df_us_da_percent.reset_index()
df_us_da_percent['job_posted_month'] = df_us_da_percent['job_posted_month_num'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_us_da_percent = df_us_da_percent.set_index('job_posted_month')
df_us_da_percent = df_us_da_percent.drop(columns='job_posted_month_num')

df_us_da_percent
```

```python
df_us_da_plot = df_us_da_percent.iloc[:, :5]

sns.lineplot(data=df_us_da_plot, dashes=False, palette='rocket_r')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Projected Job Postings')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    y_position = df_us_da_plot.iloc[-1, i]
    
    if df_us_da_plot.columns[i] == 'python':
        y_position += 2
    elif df_us_da_plot.columns[i] == 'tableau':
        y_position -= 2 
    
    plt.text(
        11.2, y_position, 
        df_us_da_plot.columns[i], 
        va='center', 
        ha='left'
    )
```
